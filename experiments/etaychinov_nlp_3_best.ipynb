## Import data

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_colwidth', 50)
pd.set_option('display.max_columns', None)

DATA_PATH = "/content/drive/MyDrive/itmo/data_nlp/matching_results.csv"
df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df['Vac_salary_from'] = pd.to_numeric(df['Vac_salary_from'].apply(lambda x: x if x != "Не указана" else np.NaN))
df['Vac_salary_to'] = pd.to_numeric(df['Vac_salary_to'].apply(lambda x: x if x != "Не указана" else np.NaN))
df.head()
```

## Catboost Training

```python
from sklearn.model_selection import train_test_split

X = df.drop(["Matching_Skills", "target"], axis=1)
y = df["target"]

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=59)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
categorical_features = df.columns.get_indexer(["Vac_exp", "Vac_schedule", "CV_exp", "CV_schedule", "CV_employment", "CV_area_name"])
text_features = df.columns.get_indexer(["Vacancy", "Employer", "Vac_description", "CV_name"]).tolist()
embeddings_features = df.columns.get_indexer(["Vac_prof_roles", "Vac_specializations", "Vac_profar_names", "Vac_keys", "CV_keys", "CV_specializations", "CV_profar_names"]).tolist()
```

```python
from catboost import Pool

train_dataset = Pool(
    X_train, y_train,
    cat_features=categorical_features,
    text_features=text_features+embeddings_features
    # embedding_features=embeddings_features
)

test_dataset = Pool(
    X_test, y_test,
    cat_features=categorical_features,
    text_features=text_features+embeddings_features
    # embedding_features=embeddings_features
)
```

```python
from catboost import CatBoostRegressor

# Попробуем покрутить гиперпараметры, depth 9 & l2_reg 0.1

model = CatBoostRegressor(
    iterations=10000,
    depth=11,
    grow_policy="Depthwise",
    l2_leaf_reg=0.1,
    task_type="GPU",
    devices='0'
)
```

```python
model.fit(train_dataset, eval_set=test_dataset)
```

```python
y_pred_train = model.predict(train_dataset)
y_pred = model.predict(test_dataset)
y_pred
```

```python
(min(y_pred_train), max(y_pred_train)), (min(y_pred), max(y_pred))
```

```python
y_pred_train[y_pred_train < 0] = 0
y_pred_train[y_pred_train > 1] = 1

y_pred[y_pred < 0] = 0
y_pred[y_pred > 1] = 1
(min(y_pred_train), max(y_pred_train)), (min(y_pred), max(y_pred))
```

```python
from sklearn.metrics import mean_absolute_error

print(f"Train MAE: {mean_absolute_error(y_train, y_pred_train)}")
print(f"Test MAE: {mean_absolute_error(y_test, y_pred)}")
```
