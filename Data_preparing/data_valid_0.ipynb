```python
import os
import csv
import pandas as pd
import xml.parsers.expat

print('Success')
```

**Чтение xml-файла весом 2гб с информацией о вакансиях и резюме**

```python
def start_element(name, attrs):
    global current_element
    current_element = name

def end_element(name):
    global current_element
    current_element = None

def char_data(char_data):
    if current_element and current_element in new_pars:
        new_pars[current_element].append(char_data.strip())

new_pars = {
    'id': [],
    'idCompany': [],
    'idCv': [],
    'idVacancy': [],
    'candidateId': [],
    'creationDate': [],
    'dateModify': [],
    'candidate': [],
    'manager': [],
    'isNew': [],
    'responseType': [],
    'regionCode': [],
    'deleted': []
}

current_element = None

p = xml.parsers.expat.ParserCreate()
p.StartElementHandler = start_element
p.EndElementHandler = end_element
p.CharacterDataHandler = char_data

with open('data/otkliki.xml', 'rb') as file:
    xml_data = file.read()
    p.Parse(xml_data)

for key, value in new_pars.items():
    print(f'{key}: {value[:5]}')
```

**Перезапись изначального xml-файла в csv-таблицу**

```python
csv_file_path = 'data/otkliki.csv'

if not os.path.exists(csv_file_path):
    with open(csv_file_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)

        # Запись заголовка
        header = list(new_pars.keys())
        writer.writerow(header)

        # Запись данных
        rows = zip(*[new_pars[key] for key in header])
        writer.writerows(rows)

    print(f'Data has been successfully saved to {csv_file_path}')
    
else:
    print(f'File {csv_file_path} already exists')
```
